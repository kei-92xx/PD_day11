## apply 메서드 활용
* apply 메서드는 사용자가 작성한 함수를 한 번에 데이터프레임의 각 행과 열에 적용하여 실행 할 수 있게 해 주는 메서드입니다.

## 데이터 프레임의 누락값을 처리한 다음 apply메서드 사용하기
### 1. 데이터프레임의 누락값 처리하기 - 열방향

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic.info()
```

```python
#count_missing 누락값의 갯수
import numpy as np

def count_missing(vec):
    null_vec = pd.isnull(vec)
    null_count = np.sum(null_vec)
    return null_count
```

```python
cmis_col = titanic.apply(count_missing)
cmis_col
```

```python
#prop_missing 누락값의 비율을 계산
#count_missing 함수를 이용하여 누락값의 개수를 구하고
#size속성을 이용하여 df의 전체 데이터 수를 구하여 나눔

def prop_missing(vec):
    num = count_missing(vec)
    dem = vec.size
    #print('누락값의 갯수(num)->%d,전체 데이터 수(dem)->%d'%(num, dem))
    return num / dem
```

```python
pmis_col = titanic.apply(prop_missing)
print(pmis_col)
```

```python
#누락값이 아닌 데이터 비율을 구함
#전체(1) - 누락값의 비율

def prop_complete(vec):
    return 1 - prop_missing(vec)
```

## 데이터프레임의 누락값을 처리하기 - 행방향

```python
cmis_row = titanic.apply(count_missing, axis=1)
pmis_row = titanic.apply(prop_missing, axis=1)
pcom_row = titanic.apply(prop_complete, axis=1)
print(cmis_row.head())
```

```python
titanic['num_missing'] = titanic.apply(count_missing, axis=1)
titanic.head()
```
